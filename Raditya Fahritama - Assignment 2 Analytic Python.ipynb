### Problem Statement

The problem states to write a program to read contents of text file and create dictionary of individual word and the frequency of the word appearing in the text file. then either display the frequencies or generate them in second file. 

###### input:

text file containing text

###### output:

frequencies of words appearing in text and textfile of it

### Solution Design

1. read the content of text file
2. preprocess the text by removing punctuation and lowercase the text
3. store the preprocessed text to file
4. read the processed file and read it word by word
5. assign the word to dictionary and set counter to plus one everytime word is found in text

###### assumption:
the text file contains text to be read


### Code

```python
#!/usr/bin/ python3

"""this code is to count word frequency in a text file"""

#importing module
import string
import json

"""this function is to preprocess the text file by removing punctuations and lowering the letters"""
def preprocess(file):
    with open(file, 'r') as fileData:
        for textline in fileData:
            #split the textfile to words
            wordsList = textline.split()
            #remove the punctuation in list comprehension
            x = [''.join(c for c in s if c not in string.punctuation) for s in wordsList]
            #lower the letters in text
            x = [x.lower() for x in x]
            #join the words per line
            y = " ".join(x)
            #write to textfile
            fila = open(prefile, "a")
            fila.write(y+"\n")
            
"""this function is to clear the processed text file"""
def cleartxt():
    f = open(prefile, "w")
    f.write("")

#delcaring text file to be read and text file to store the processed text
filename = 'BohemianRhapsody.txt'
prefile = 'preprocessed.txt'

#delcare empty dictionary
dictionary = {}

#run preprocess function
preprocess(filename)

"""this part is to read the words and store it to dictionary"""
with open(prefile, "r") as f:
    #read by words
    words = f.read().split()
    #convert to list
    words = list(words)
    #sort the word alphabetically
    words = sorted(words)
    #looping by word and assign word to dictionary and add the value counter
    for x in words:
        if x in dictionary:
            dictionary[x] += 1
        else:
            dictionary[x] = 1
            
#display the results 
print(open(filename, "r").read(),'\n')
print("--- DICTIONARY OF WORDS ---\n",dictionary)

#store the frequency dictionary to file
with open('wordfreq.txt', 'w') as convert_file:
     convert_file.write(json.dumps(dictionary))

#run the clear text function
cleartxt()
```

### Problem Statement
The problem states to write program that read contents of file, and store the word in dictionary for every word found and in which lines that we can find the word from the text file. then we store the dictionary in file.   

###### input:
text file to be read

###### output:
dictionary of line index in which words appearing and textfile of it


### Solution Design

1. read the content of text file
2. preprocess the text by removing punctuation and lowercase the text
3. store the preprocessed text to file
4. read the processed file and read words one by one
5. make function to enumerate line in file. and if the word that is currently in read is found, append the line counter to list
6. return the word and counter and append them in dictionary

###### assumption:
the text file contains valid text inside


### Code

```python
#!/usr/bin/ python3

"""this code is to assign word and line index of the word from a text file to a dictionary"""

#import required module
import string
import json

"""this function is to preprocess the text file by removing punctuations and lowering the letters"""
def preprocess(file):
    with open(file, 'r') as fileData:
        for textline in fileData:
            #split the textfile to words
            wordsList = textline.split()
            #remove the punctuation in list comprehension
            x = [''.join(c for c in s if c not in string.punctuation) for s in wordsList]
            #lower the letters in text
            x = [x.lower() for x in x]
            #join the words per line
            y = " ".join(x)
            #write to textfile
            fila = open(prefile, "a")
            fila.write(y+"\n")

"""this function is to get line index of the word"""
def lineindex(filepath, keywords):
    with open(filepath) as f:
        #assign empty list
        countlist = []
        #loop for every file line. if the find function find a word in the line, append the counter to list
        for counter, line in enumerate(f, start = 1):
            if line.find(keywords) >= 0:
                countlist.append(counter)
        #return the word and the list
        return(keywords, countlist)

"""this function is to clear the processed text file"""
def cleartxt():
    f = open(prefile, "w")
    f.write("")
    
#delcaring text file to be read and text file to store the processed text
filename = 'BohemianRhapsody.txt'
prefile = 'preprocessed.txt'

#delcare empty dictionary
dictionary = {}

#run preprocess function
preprocess(filename)

"""this part is to read the processed file to get the words and run the line indexing function"""
with open(prefile, "r") as f:
    #read the file by words
    words = f.read().split()
    #set the list so that the list doesn't contain duplicate
    sets = list(set(words))
    #sort the list
    sets = sorted(sets)
    #looping for every word in the list
    for x in sets:
        if len(x) > 1 :
            key,value = lineindex(prefile, x)
            dictionary[key] = value

#display the results             
print(open(filename, "r").read(),'\n')
print("--- DICTIONARY OF WORDS ---\n",dictionary)

#store the frequency dictionary to file
with open('wordline.txt', 'w') as convert_file:
     convert_file.write(json.dumps(dictionary))

#run the clear processed file
cleartxt()
```

```python
#!/usr/bin/ python3
"""this code is to create class and subclass and create object for the sublass based on user input"""

"""this class is the employee superclass like in problem 3"""
class Employee:
    #initial attributes declaration
    def __init__(self, name, idnum, department, jobtitle):
        self.__name = name
        self.__idnum = idnum
        self.__department = department
        self.__jobtitle = jobtitle
    #setter function for every attribute
    def set_name(self, name):
        self.__name = name
    def set_idnum(self, idnum):
        self.__idnum = idnum
    def set_department(self, department):
        self.__department = department
    def set_jobtitle(self, jobtitle):
        self.__jobtitle = jobtitle
    #getter function for every attribute
    def get_name(self):
        return self.__name 
    def get_idnum(self):
        return self.__idnum
    def get_department(self):
        return self.__department
    def get_jobtitle(self):
        return self.__jobtitle

"""this class is the subclass of Employee superclass"""
class ProductionWorker(Employee):
    #initial attributes declaration
    def __init__(self, name, idnum, department, jobtitle, shiftnum, hourlypay):
        self.__shiftnum = shiftnum
        self.__hourlypay = hourlypay
        Employee.__init__(self, name, idnum, department, jobtitle)
    #setter function for every attribute
    def set_shiftnum(self, shiftnum):
        self.__shiftnum = shiftnum
    def set_hourlypay(self, hourlypay):
        self.__hourlypay = hourlypay
    #getter function for every attribute
    def get_shiftnum(self):
        return self.__shiftnum 
    def get_hourlypay(self):
        return self.__hourlypay

"""this part is to get user's input for creating the object"""
loop = 1
while loop == 1:
    try:
        #getting user's inputs
        ename = input("Input employee name : ")
        eid = int(input("Input employee ID : "))
        edepartment = input("Input department : ")
        etitle = input("Input job title : ")
        #getting shift with simple input checker
        loop2 = 1
        while loop2 == 1:
            eshift = int(input("Input shift of employee (1 for day, 2 for night) : "))
            if eshift != 1 and eshift != 2: 
                print("\nPlease input either 1 or 2!\n")
            else:
                loop2 = 0
        epay = int(input("Input employee hourly rate : "))
        
        #check if input is blank for every input
        if ename == "" or eid == "" or edepartment == "" or etitle == "" or eshift == "" or epay == "":
            print("\nPlease don't leave blank on input!\n")
        else:
            loop = 0
    except:
        #excpetion handling
        print("Invalid input!")
            
#create object of subclass
pr = ProductionWorker(ename, eid, edepartment, etitle, eshift, epay)

"""this part is to display object with getter method"""
#display object with getter function from both subclass and superclass
print("\n--Employee Profile--\n")
print("Name : ",pr.get_name())
print("ID : ",pr.get_idnum())
print("Department : ",pr.get_department())
print("Job title : ",pr.get_jobtitle())
print("Shift : ",pr.get_shiftnum())
print("Hourly pay : $",pr.get_hourlypay())
```

- If m = 0, then return n+1
- If n = 0, then Ackermann(m-1, n)
- Otherwise, return Ackermann(m-1, Ackermann(m, n-1))
then, we test it with small values

###### input:

ackermann's function result


### Solution Design
1. define ackermann's recursive function as following:
    - If m = 0, then return n+1
    - If n = 0, then Ackermann(m-1, n)
    - Otherwise, return Ackermann(m-1, Ackermann(m, n-1))

###### assumption:
-

### Code

```python
#!/usr/bin/ python3
"""this code is to define and execute ackermann's function"""

"""this function is the ackermann's recursive function logic based on the problem statement"""
def ackermann(m, n):
    print(m,n)
    if m == 0:
        #return n + 1 if m == 0
        return n + 1
    elif n == 0:
        #return recursive with m-1 and n as parameter
        return ackermann(m - 1, n)
    else:
        #return recursive with m - 1 and another recursive m and n-1 as parameter
        return ackermann(m - 1, ackermann(m,n-1))

#run the program
ackermann(2, 3)
 
```
